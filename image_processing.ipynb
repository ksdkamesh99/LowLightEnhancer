```
image_high=[]
image_low=[]
i=0
for img_name in list_high:
  i=i+1
  img_high=Image.open("/content/drive/My Drive/LowLightEnhancement/Shared Dataset Files/Dataset/high/"+img_name)
  # print(img_high)
  img_arr_high=np.asarray(img_high)
  img_low=Image.open("/content/drive/My Drive/LowLightEnhancement/Shared Dataset Files/Dataset/low/"+img_name)
  img_arr_low=np.asarray(img_low)
  img_arr_high=cv2.resize(img_arr_high,(96,96))
  img_arr_low=cv2.resize(img_arr_low,(96,96))
  # plt.imshow(img_arr_low)
  # plt.show()
  image_high.append(img_arr_high)
  image_low.append(img_arr_low)
  print(i)
  
  
```
